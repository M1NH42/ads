```python
plt.plot(3, 2, '.')
```

```python
from matplotlib.backends.backend_agg import FigureCanvasAgg
from matplotlib.figure import Figure

#create new fig
fig = Figure()

# associate fig with the backend
canvas = FigureCanvasAgg(fig)

# add a subplot to the fig
ax= fig.add_subplot(111)

# plot the point (3,2)
ax.plot(3, 2, '.')
```

```python
# create a new figure
plt.figure()

# plot the point (1.5, 1.5) using the circular marker
plt.plot(1.5, 1.5, 'o')

# plot the point (2, 2) using the circular marker
plt.plot(2, 2, 'o')

# plot the point (2.5, 2.5) using the circular marker
plt.plot(2.5, 2.5, 'o')
```
