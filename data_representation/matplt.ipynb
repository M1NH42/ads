```python
# The usage of the index_col and parse_dates parameters of the read_csv function to define 
#the first (0th) column as index of the resulting DataFrame and convert the dates in the 
#column to Timestamp objects, respectively.
air_quality = pd.read_csv('data.csv', index_col=0, parse_dates=True)

air_quality.head()
```

```python
air_quality.plot()
```

```python
air_quality['station_paris'].plot()
```

```python
air_quality['station_london'].plot()
```

```python
# compare the NO2 values measured in london vs paris
air_quality.plot.scatter(x='station_london', y = 'station_paris', alpha = .7)
```

```python
[method_name for method_name in dir(air_quality.plot) if not method_name.startswith('_')]
```

```python
# boxplot method
air_quality.plot.box()
```

```python
axs = air_quality.plot.area(figsize=(12, 4), subplots= True)
```

```python
fig, axs = plt.subplots(figsize = (12, 4));
air_quality.plot.area(ax=axs);
axs.set_ylabel('NO$_2$ concentration');
fig.savefig('no2_concentration.png')
```

```python
# add dataset
titanic = pd.read_csv('titanic.csv')
```

```python
titanic.head()
```

```python
# calculate summary statistics

# average age of titanic passengers
titanic['Age'].mean()
```

```python
titanic[['Age', 'Fare']].median()
```

```python
titanic[['Age', 'Fare']].describe()
```

```python
titanic.agg({'Age': ['min', 'max', 'mean', 'median'], 'Fare': ['min', 'max', 'mean', 'median']})
```

```python
# aggreagting statistics grouped by category
#  the groupby() method is applied on the Sex column to make a group per category
titanic[['Sex', 'Age']].groupby('Sex').mean()
```

```python
titanic.groupby('Sex').mean()
```

```python
titanic.groupby('Sex')['Age'].mean()
```

```python
titanic.groupby(['Sex', 'Pclass'])['Fare'].mean()
```

```python
# Count number of records by category
titanic['Pclass'].value_counts()
```

```python
len(titanic.groupby(['Age', 'Pclass'])['Fare'].mean())
```

```python
titanic.groupby('Pclass')['Pclass'].count()
```
